```python
import cv2
import os
import matplotlib.pyplot as plt
import random
```

```python
# Random visualization
DATA_PATH = "../processed_data/train"
images = os.listdir(f"{DATA_PATH}/images")

sample = random.choice(images)

img = cv2.imread(f"{DATA_PATH}/images/{sample}",0)
mask = cv2.imread(f"{DATA_PATH}/masks/{sample}",0)

plt.imshow(img, cmap = "gray")
plt.imshow(mask,alpha = 0.5)
plt.title(sample)
plt.show()
```

```python
import numpy as np

unique = np.unique(mask)
print("Mask values:", unique)
```

```python
# 5-10 random samples check 
for _ in range(5):
    sample = random.choice(images)
    img = cv2.imread(f"{DATA_PATH}/images/{sample}",0)
    mask = cv2.imread(f"{DATA_PATH}/masks/{sample}",0)

    plt.imshow(img, cmap = "gray")
    plt.imshow(mask,alpha = 0.5)
    plt.title(sample)
    plt.show()
```
